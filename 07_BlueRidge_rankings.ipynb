```python
def br_importer(df, year):
    """
    Function to import and clean blueridge specialty-specific rankings
    Args:
        df (dataframe)
        year (int)
    """
    df.columns = df.columns.str.title()
    df["Organization Name"] = df["Organization Name"].str.lower() 
    df = df.dropna(subset=["Rank"])
    cols = ["Organization Name", "Anesthesiology", "Dermatology", 'Emergency Medicine', 'Family Medicine', 'Genetics', 'Internal Medicine/Medicine', 'Neurology', 'Neurosurgery', 'Obstetrics & Gynecology', 'Ophthalmology', 'Orthopedics', 'Otolaryngology', 'Pathology', 'Pediatrics', 'Physical Medicine & Rehab', 'Psychiatry', 'Public Health & Prev Medicine', 'Radiation-Diagnostic/Oncology', "Surgery", "Urology"]
    df = df[cols]
    specialties = [c for c in cols if c != "Organization Name"]
    df = df.melt(id_vars=["Organization Name"], value_vars=specialties, var_name="specialty", value_name="blueridge_money")
    renamed_specialties = {"Genetics": "Pediatrics-Medical Genetics", 'Internal Medicine/Medicine': "Internal Medicine", "Neurosurgery": "Neurological Surgery", "Obstetrics & Gynecology": "Obstetrics and Gynecology", "Orthopedics": "Orthopaedic Surgery", "Physical Medicine & Rehab": "Physical Medicine and Rehabilitation", "Public Health & Prev Medicine": "Preventive Medicine", "Radiation-Diagnostic/Oncology": 'Radiation Oncology'}  #not for radiation-diagnostic and oncology!!!
    df["specialty"] = df["specialty"].replace(renamed_specialties)

    br_list = []
    for spec in df["specialty"].unique():
        br_temp = df[df["specialty"]==spec].copy()
        br_temp = br_temp.dropna(subset=["blueridge_money"])
        br_temp = br_temp.sort_values(by="blueridge_money", ascending=False).reset_index(drop=True)
        br_temp["blueridge_rank"] = np.arange(1, len(br_temp)+1)
        br_temp["year"] = year
        br_list.append(br_temp)
    df = pd.concat(br_list, ignore_index=True)
    return df
```
